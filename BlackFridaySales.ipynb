```python
import category_encoders as ce

#encode categorical variables with ordinal encoding
encoder = ce.OrdinalEncoder(cols = ['Gender', 'City_Category','Age','Stay_In_Current_City_Years'])

X_train = encoder.fit_transform(X_train)
X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
#Import Random Forest Regressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error

#Instantiate Regressor
# rfc = RandomForestRegressor(n_estimators = 400, max_depth = 15, min_samples_split= 2, random_state = 0)
model_xgb = xgb.XGBRegressor(n_estimators = 100, max_depth = 7, min_samples_split= 2)


#Fit the model
model_xgb.fit(X_train, y_train)

#Predict the test set results
y_pred = model_xgb.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print('Mean Squared Error:', mse)
print('Mean Absolute Error:', mae)
```
